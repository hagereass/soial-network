```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from community import community_louvain
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
import random
import urllib.request
```

```python
#1. DOWNLOAD & LOAD FACEBOOK GRAPH (EGONET)
urllib.request.urlretrieve(
    "https://snap.stanford.edu/data/facebook_combined.txt.gz",
    "fb.txt.gz"
)

# Load graph
G = nx.read_edgelist("fb.txt.gz", create_using=nx.Graph())

print("Graph Loaded")
print("Nodes:", G.number_of_nodes())
print("Edges:", G.number_of_edges())
```

```python
# 2. GRAPH METRICS



degree_dict = dict(G.degree())
clustering_dict = nx.clustering(G)
betweenness_dict = nx.betweenness_centrality(G)
communities = community_louvain.best_partition(G)


features_df = pd.DataFrame({
'node': list(G.nodes()),
'degree': [degree_dict[n] for n in G.nodes()],
'clustering': [clustering_dict[n] for n in G.nodes()],
'betweenness': [betweenness_dict[n] for n in G.nodes()],
'community': [communities[n] for n in G.nodes()],
})

features_df.head()
```

```python
# 3. BASELINE BOT LABELING
# =============================
random.seed(42)
nodes = list(G.nodes())
bot_count = int(0.1 * len(nodes))
bots = set(random.sample(nodes, bot_count))


features_df['label'] = features_df['node'].apply(lambda x: 1 if x in bots else 0)
```

```python
# 4. BASELINE CLASSIFIER
# =============================
X = features_df[['degree', 'clustering', 'betweenness', 'community']]
y = features_df['label']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
clf = RandomForestClassifier(n_estimators=300, random_state=42)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)


print("===== BASELINE PERFORMANCE =====")
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
# 5. STRUCTURAL EVASION ATTACK
# =============================
G_evasion = G.copy()
top_nodes = sorted(degree_dict, key=degree_dict.get, reverse=True)[:8]


for b in bots:
    for t in top_nodes:
        if b != t:
            G_evasion.add_edge(b, t)

# Recalculate metrics
degree_ev = dict(G_evasion.degree())
clustering_ev = nx.clustering(G_evasion)
betweenness_ev = nx.betweenness_centrality(G_evasion)
communities_ev = community_louvain.best_partition(G_evasion)


features_ev = pd.DataFrame({
    'node': list(G_evasion.nodes()),
    'degree': [degree_ev[n] for n in G_evasion.nodes()],
    'clustering': [clustering_ev[n] for n in G_evasion.nodes()],
    'betweenness': [betweenness_ev[n] for n in G_evasion.nodes()],
    'community': [communities_ev[n] for n in G_evasion.nodes()],
})


features_ev['label'] = features_ev['node'].apply(lambda x: 1 if x in bots else 0)


X_ev = features_ev[['degree', 'clustering', 'betweenness', 'community']]
y_ev = features_ev['label']


X_train_ev, X_test_ev, y_train_ev, y_test_ev = train_test_split(X_ev, y_ev, test_size=0.3, random_state=42)
clf_ev = RandomForestClassifier(n_estimators=300, random_state=42)
clf_ev.fit(X_train_ev, y_train_ev)
y_pred_ev = clf_ev.predict(X_test_ev)


print("===== AFTER STRUCTURAL EVASION =====")
print("Accuracy:", accuracy_score(y_test_ev, y_pred_ev))
print(classification_report(y_test_ev, y_pred_ev))
```

```python
# 6. GRAPH POISONING ATTACK

G_poison = G.copy()
# Convert node IDs to integers to find the true maximum numerical value
max_node_int = max(int(n) for n in nodes)
new_nodes = range(max_node_int + 1, max_node_int + 1 + bot_count)


for n in new_nodes:
      targets = random.sample(nodes, 5) # connect to 5 real nodes
      for t in targets:
          G_poison.add_edge(n, t)

# Recompute metrics after poisoning
degree_p = dict(G_poison.degree())
clustering_p = nx.clustering(G_poison)
betweenness_p = nx.betweenness_centrality(G_poison)
communities_p = community_louvain.best_partition(G_poison)


features_p = pd.DataFrame({
'node': list(G_poison.nodes()),
'degree': [degree_p[n] for n in G_poison.nodes()],
'clustering': [clustering_p[n] for n in G_poison.nodes()],
'betweenness': [betweenness_p[n] for n in G_poison.nodes()],
'community': [communities_p[n] for n in G_poison.nodes()],
})
# Label: old bots + new fake nodes
features_p['label'] = features_p['node'].apply(lambda x: 1 if x in bots or x in new_nodes else 0)


X_p = features_p[['degree', 'clustering', 'betweenness', 'community']]
y_p = features_p['label']


X_train_p, X_test_p, y_train_p, y_test_p = train_test_split(X_p, y_p, test_size=0.3, random_state=42)
clf_p = RandomForestClassifier(n_estimators=300, random_state=42)
clf_p.fit(X_train_p, y_train_p)
y_pred_p = clf_p.predict(X_test_p)


print(" AFTER GRAPH POISONING ")
print("Accuracy:", accuracy_score(y_test_p, y_pred_p))
print(classification_report(y_test_p, y_pred_p))
```

```python
# 7. VISUALIZATION

plt.figure(figsize=(7, 7))
nx.draw(G, node_size=10)
plt.title("Original Graph")
plt.show()


plt.figure(figsize=(7, 7))
nx.draw(G_evasion, node_size=10)
plt.title("Structural Evasion Attack")
plt.show()


plt.figure(figsize=(7, 7))
nx.draw(G_poison, node_size=10)
plt.title("Graph Poisoning Attack")
plt.show()
```

```python
# 8. TEST POISONED MODEL ON CLEAN DATA

X_clean = X
y_clean = y
clean_pred = clf_p.predict(X_clean)
print(" POISONED MODEL TESTED ON CLEAN GRAPH ")
print("Accuracy:", accuracy_score(y_clean, clean_pred))
print(classification_report(y_clean, clean_pred))
```

```python
# 9. PERFORMANCE COMPARISON TABLE



results = pd.DataFrame({
'Condition': ['Baseline', 'Structural Evasion', 'Graph Poisoning', 'PoisonedClean Test'],
'Accuracy': [
accuracy_score(y_test, y_pred),
accuracy_score(y_test_ev, y_pred_ev),
accuracy_score(y_test_p, y_pred_p),
accuracy_score(y_clean, clean_pred)
]
})
print(" PERFORMANCE COMPARISON")
print(results)
```

```python
# 10. ACCURACY BAR PLOT

plt.figure(figsize=(7,5))
plt.bar(results['Condition'], results['Accuracy'])
plt.ylabel('Accuracy')
plt.title('Bot Detection Accuracy Across All Scenarios')
plt.xticks(rotation=20)
plt.show()
```
